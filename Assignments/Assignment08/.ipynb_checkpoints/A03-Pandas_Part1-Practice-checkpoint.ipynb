#### Task 5c: Make selections with `loc`

+ Select a single item with `at`.
+ Select a row slice with `loc`.
+ Select a row and column slice with `loc`.

```python
#.iat() # 
sub10 = iris_df[::10] # using every 10th row from iris.csv file
sub10

```

```python
#### Task 5d: Make selections with `iloc` and `iat`.
### at and iat a meant to access a scalar, that is, a single element in the dataframe, while loc and ilocare ments to access several elements at the same time

+ Select a single item with `iat`.
+ Select a row slice with `iloc`.
+ Select a row and column slice with `iloc`.
```

```python
sub10.iat[10,3] # in this new dataframe, row with index 10 is 90, column with index 3 is still petal_length
```

```python
#.at() # considers the raw labels; even integer indexes are treated as labels
# syntax: sub10.iat('row label', 'column label of cell')

sub10.at[10, 'petal_width'] # will look for the row with label 10
```

```python
# .loc[] is a label based method. 
# my_dataframe.loc["row label"] 

sub10.loc[10] # will also return row with label 10
```

```python
# .iloc[] takes slices based on index’s position
# syntax: mydata_frame.iloc[positional index number]
          # mydata_frame.iloc[0:3]

# Suppose you want to get the first three elements of the data set.
# sub10.iloc[10, 'petal_width'] # returns error msg; second parameter is also a positional index
sub10.iloc[10, 4]
```

```python
sub10.loc[0, :] # first row, all the columns
```

```python
sub10.loc[[0, 10, 20], :] # first row (label 0) through row with label 20, all the columns
```

```python
sub10.loc[0:20, :] # first row (label 0) through row with label 20, all the columns
```

## Select a row and column slice with `iloc`
# syntax: my_dataframe.iloc[row slicing, column slicing]


```python
sub10.iloc[0:2, 1:3]
```

#### Task 5e: Boolean Indexing

Create subsets using boolean indexes that:
+ Use one boolean operator.
+ Use two boolean operators.


```python
sub10[sub10.species == 'setosa']
```

```python
sub10[sub10.species != 'setosa']
```

```python
sub10[sub10.sepal_length > 5.0]
```

```python
sub10[(sub10.sepal_length > 5.0) & (sub10.sepal_length > 5.0)]
```
